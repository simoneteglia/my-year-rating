```python
# Importa librerie
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
from math import pi
```

```python
# Dati dei voti [2021, 2022]
voti = {
    'Pier': [8, 4],
    'Clara': [5, 7],
    'SdG': [4, 6.75],
    'Belmy': [2, 5],
    'Sofia': [6.5, 7.75],
    'Valeria': [7.5, 7.5],
    'Clarke': [7, 8.5],
    'Simo': [7, 8.5],
    'Dudu': [3.5, 5.5],
    'Jacopo': [5, 6.5],
    'Ludona': [2, 6],
    'Luisa': [8, 8],
    'Sere': [7, 8],
    'Michele': [8, 7.25],
    'Gaia': [8, 9]
}
```

```python
# Crea il DataFrame
df = pd.DataFrame(voti, index=['2021', '2022'])

# Mostra il DataFrame
print(df)
```

```python
# Trasponi il DataFrame
df_t = df.T
```

```python
# Media
media = df_t.mean()
print(media)
```

```python
# Moda
moda = df_t.mode()
print(moda)
```

```python
# Mediana
mediana = df_t.median()
print(mediana)
```

```python
# Varianza
varianza = df_t.var()
print(varianza)
```

```python
# Deviazione standard
dev_std = df_t.std()
print(dev_std)
```

```python
# Crea una figura con 5 sottografici
fig, ax = plt.subplots(5, 1, sharex=True, figsize=(8, 16))

# Crea il grafico a linee per la media
ax[0].plot(df_t.mean())
ax[0].set_ylabel('Media')

# Crea il grafico a linee per la moda
ax[1].plot(df_t.mode().iloc[0])
ax[1].set_ylabel('Moda')

# Crea il grafico a linee per la mediana
ax[2].plot(df_t.median())
ax[2].set_ylabel('Mediana')

# Crea il grafico a linee per la varianza
ax[3].plot(df_t.var())
ax[3].set_ylabel('Varianza')

# Crea il grafico a linee per la deviazione standard
ax[4].plot(df_t.std())
ax[4].set_ylabel('Dev. standard')

# Mostra la figura
plt.show()
```

```python
def aaa(voti_anno, colore, bordo, anno):
  frequenze = {}
  for voto in voti_anno:
    if voto in frequenze.keys():
      frequenze[voto] += 1
    else:
      frequenze[voto] = 1
  print(frequenze)

  plt.figure(figsize=(16, 8))
  plt.bar(frequenze.keys(),frequenze.values(), color=colore, edgecolor=bordo, width=0.2)
  plt.xticks(list(frequenze.keys()))
  plt.legend([anno], loc="upper left", bbox_to_anchor=(1,1))
  plt.show()
```

```python
# Funzione per creare il grafico a radar
def radar(stats, title):
    # Set dei valori
    values = stats.values

    # Numero di variabili
    categories = list(stats.index)
    N = len(categories)

    # Aggiungi un valore fittizio alla fine del set di valori
    values = np.append(values, values[0])

    # Angoli di divisione delle sezioni
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    # Inizializza il grafico
    ax = plt.subplot(111, polar=True)

    # Disegna le linee di riferimento
    plt.xticks(angles[:-1], categories, color='grey', size=12)
    ax.set_rlabel_position(0)
    plt.yticks([5, 10, 15], ["5", "10", "15"], color="grey", size=7)
    plt.ylim(0, 15)

    # Disegna le barre
    ax.plot(angles, values, linewidth=1, linestyle='solid')
    ax.fill(angles, values, 'b', alpha=0.1)
    ax.set_rmax(10)

    # Imposta il titolo
    plt.title(title, size=11, y=1.1)

# Crea il grafico a radar per ogni persona
for col in df_t.columns:
    radar(df_t[col], col)
    plt.show()
```

```python
# Crea il grafico a radar per ogni persona
fig, axs = plt.subplots(1, 2, figsize=(10, 5), sharey=True)
titles = df_t.columns

for ax, col, title in zip(axs, df_t.columns, titles):
    radar(df_t[col], title)
    ax.title.set_text(title)

# Aggiungi la legenda
plt.legend(titles, loc="upper left", bbox_to_anchor=(1,1))

plt.show()
```
